```python
from pandas.io.json import json_normalize

results = response.json()
df = json_normalize(results)
df.info()
df
```

```python
df['category'] = df['category'].astype('category')
df['price'] = df['price'].astype('float')
```

```python
df.info()
```

```python
## sacar el precio medio segun la catagoria y teniendo que estar en la misma divisa 
```

```python
avg_mpg = df.groupby(['category', 'currency'], as_index=False).agg({'price':'mean'})

print(avg_mpg)
```
