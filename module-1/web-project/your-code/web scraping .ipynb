```python
##Name of the most recently added 

from bs4 import BeautifulSoup

html = requests.get(url).content
soup = BeautifulSoup(html, "lxml")
soup
```

```python
##HEADER 

tags = ['h1', 'h2', 'h3', 'h4', 'h5', 'h6', 'h7', 'p']
text = [element.text for element in soup.find_all(tags)]
text[0]
```

```python
#LINK OF IMAGES
images = soup.select(".image")
images[0].attrs["href"]
    
```

```python
##FROM QUEENN´S PAGE TAKE THE DATA FROM THE SONGS 
url_q = "https://es.wikipedia.org/wiki/Queen_(%C3%A1lbum)"
html_q = requests.get(url_q).content

soup_q = BeautifulSoup(html_q, "lxml")
filtrado = soup_q.select(".tracklist [href]")
songs= [i.text.strip()for i in filtrado]
songs
```
